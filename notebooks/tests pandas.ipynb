### 4. Extraire l’information concernant l’année de plantation pour l’ensemble des lignes (rows) et stocker ces informations dans une autre variable de type List

```python
def index_colonne(data, nom_colonne):
    i = afficher_ligne(data, 0)
    for elm in i:
        if str(elm) == nom_colonne:
            return i.index(elm)

# Donne l'index de la colonne "nom_colonne" de "fichier_csv"
# "anneedeplantation" ==> index 19
# "genre_bota" ==> index 13
# "especes" ==> index 14

def extraire_all_colonne(data, nom_colonne):
    index_c = int(index_colonne(data, nom_colonne))
    liste_annee = [row[index_c] for row in data[1:]]
    return liste_annee

# Renvoie une liste des elements de la colonne "nom_colonne" du "fichier_csv"
    
print(f"l'index de l'année de plantation est {index_colonne(data, 'anneedeplantation')}")
print(extraire_all_colonne(data, "anneedeplantation"))
```

```python
def afficher_50_premières_dernières(data, nom_colonne):
    liste = extraire_all_colonne(data, nom_colonne)
    l2 = liste[1:51] + liste[-50:]
    return l2

# Affiche les 50 premiers & 50 derniers elements de la colonne "nom_colonne"

print(afficher_50_premières_dernières(data, "anneedeplantation"))
```

### 9. Quelle est la plus ancienne année de plantation recensée dans ce dataset ? La plus récente ?    

```python
def date_ancienne_recente(data):
    liste = extraire_all_colonne(data, "anneedeplantation")
    ancienne = int(liste[0])
    recente = int(liste[0])
    for elm in liste:
            if elm != "":
                if int(elm) < int(ancienne):
                    ancienne = elm
                if int(elm) > int(recente):
                    recente = elm
    return ancienne, recente

# Retourne la date de plantation la plus ancienne et la plus récente des arbres référencés dans "fichier_csv"

dateA_R = date_ancienne_recente(data)
print(dateA_R)
print(f"La date de plantation la plus ancienne est {dateA_R[0]} et la plus récente est {dateA_R[1]}.")
```

### 10. Combien d’arbres ont été plantés année par année (ex : 1987 : 771, 1988 : 266, etc…) ?

```python
def elm_par_colonne(data, nom_colonne):
    liste = extraire_all_colonne(data, nom_colonne)
    dico = {}
    for elm in liste:
        if elm not in dico:
            dico[elm] = 1
        else:
            dico[elm] += 1
    return dico

# Comptabilise le nombre d'élements par critères uniques de la colonne "nom_colonne".

print(elm_par_colonne(data, 'anneedeplantation'))
```

### 11. Combien d’arbres ont été plantés en moyenne chaque année ?   

```python
def moyenne_par_colonne(data, nom_colonne):
    dico = elm_par_colonne(data, nom_colonne)
    liste = [value for value in dico.values()]
    moyenne = int(sum(liste)/len(liste))
    return moyenne

# Donne la moyenne d'arbre plantés chaque année.

print(f"En moyenne, {moyenne_par_colonne(data, 'anneedeplantation')} arbres sont plantés par ans.")
```

### 12. Stocker conjointement l’année de plantation et le nombre d’arbres plantés dans un tuple. Les tuples seront stockés dans une liste.

```python
def liste_tuple(data, nom_colonne):
    dico = elm_par_colonne(data, nom_colonne)
    liste = [(key, value) for key, value in dico.items()]
    tri = sorted(liste, key = lambda x: x[0], reverse = True)
    return tri

# Renvoie une liste composée de tuples ("anneedeplantation", "Nbr_arbre_plantés")

print(liste_tuple(data,'anneedeplantation'))
```

### 3. En utilisant la structure de données Dict compter le nombre de représentants de chaque genre botanique.

```python
def nbr_elem_par_colonne(data, nom_colonne):
    liste = extraire_all_colonne(data, nom_colonne)
    dico = {}
    for elm in liste:
        if elm not in dico:
            dico[elm] = 1
        else:
            dico[elm] += 1
    return dico

# Comptabilise le nombre d'arbres pour chaque éléements différents de la colonne "nom_colonne" sous forme de dictionnaire

print(nbr_elem_par_colonne(data, "genre_bota"))
```

### 4. Quel est le genre botanique le plus représenté ?

```python
def le_plus_represente(data, nom_colonne):
    dico = nbr_elem_par_colonne(data, nom_colonne)
    genre = ""
    nbr = 0
    for key in dico:
        value = dico[key]
        if value > nbr:
            genre = key
            nbr = value
    return genre, nbr

# Permet de connaitre l'élement le plus représenté de la colonne "nom_colonne" et son nombre de représentants.

rep_genre = le_plus_represente(data, "genre_bota")

print(f"Le genre botanique le plus représenté est {rep_genre[0]} avec {rep_genre[1]} individus.")
print(rep_genre)
```

### 5. Classer tous les genres botaniques en fonction de leur nombre (ordre décroissant).

```python
def tri_decroissant_nombre_elm(data, nom_colonne):
    dico = nbr_elem_par_colonne(data, nom_colonne)
    tri = dict(sorted(dico.items(), key = lambda item: item[1], reverse = True))
    return tri

# Retourne le dictionnaire ordonné de façon décroissante de représentant de "nom_colonne".

print(tri_decroissant_nombre_elm(data, "genre_bota"))
```

### 6. Quelle est la proportion du genre botanique le plus représenté parmi tous les genres ?
###    Quelle est la proportion de chacun des genres botaniques par rapport au total ?
###    Idem précédemment mais en comptant les espèces. 

```python
def tri_proportion(data, nom_colonne):
    dico = tri_decroissant_nombre_elm(data, nom_colonne)
    for key in dico:
        value = dico[key]
        dico[key] = round((value*100/31826), 2)
    return dico

# Retourne le dictionnaire de la proportion de chaque élement de "nom_colonne" par rapport au total, par ordre décroissant.

print(tri_proportion(data, "genre_bota"))
print("\n")
print(tri_proportion(data, "espece"))
```

### 7. Comment la pratique de plantation a-t-elle évolué au cours du temps ? Diversifie-t-on plus maintenant qu’avant ?

```python
def elm_par_annee(fichier_csv):
    liste = tri_genre_annee(fichier_csv)
    dico = {}
    reste = []
    for elm in liste:
        if elm[0] not in dico:
            dico [elm[0]] = 1
            reste.append(elm)
        else:
            if elm not in reste:
                dico[elm[0]] += 1
            else:
                continue
    tri = dict(sorted(dico.items(), key = lambda item: item[1], reverse = True))
    return tri

# Retourne le dictionnaire (ordonné décroissant) du nombre de genre botanique différents planté par année

print(elm_par_annee('../data/trees.csv'))
```
